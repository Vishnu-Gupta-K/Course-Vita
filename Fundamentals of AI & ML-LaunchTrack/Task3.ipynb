## Part 1: Understanding our Data (Descriptive Statistics)

We have a file called `House Prices - Advanced Regression Techniques.csv`. We want to answer three simple questions about the `SalePrice` column:
1.  **Mean:** What is the average price?
2.  **Median:** If we line up all the houses from cheapest to most expensive, what is the price of the one in the exact middle?
3.  **Mode:** What is the most common price?


```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/House Prices - Advanced Regression Techniques.csv')
prices = df['SalePrice']

average_price = prices.mean()   # The Mean
middle_price = prices.median()  # The Median
most_common = prices.mode()[0]  # The Mode

print("--- House Price Statistics ---")
print("Mean (Average):   $", round(average_price, 2))
print("Median (Middle):  $", round(middle_price, 2))
print("Mode (Most Common): $", round(most_common, 2))
```

```python
plt.figure(figsize=(12, 5))

# CHART 1: Histogram
plt.subplot(1, 2, 1)  # 1 row, 2 columns, this is the 1st chart
plt.hist(prices, bins=30, color='skyblue', edgecolor='black')
plt.title('How many houses at each price?')
plt.xlabel('Price ($)')
plt.ylabel('Count of Houses')

# CHART 2: Box Plot
plt.subplot(1, 2, 2)
plt.boxplot(prices)
plt.title('Range of Prices')
plt.ylabel('Price ($)')

plt.show()
```

## Part 2: The "Coin Toss" Experiment

If you flip a coin, the chance of getting Heads is 50% (0.5).
But if you flip it only 5 times, you might get Heads 4 times (80%). That doesn't mean the coin is broken!

We will write a **Loop** to flip a coin 1,000 times. We expect that as we flip it more and more, the average number of Heads will get closer and closer to 50%.

```python
import random
number_of_tosses = 1000
heads_count = 0
probabilities = []  # We will store the results here


for i in range(1, number_of_tosses + 1):
    flip = random.choice([0, 1]) # flip: 0 is Tails, 1 is Heads
    if flip == 1:
        heads_count = heads_count + 1
    current_avg = heads_count / i
    probabilities.append(current_avg)

plt.figure(figsize=(10, 5))
plt.plot(probabilities)
plt.axhline(0.5, color='red', linestyle='--', label='Target (0.5)')
plt.title('Coin Toss Predictor: Watching the Probability Settle')
plt.xlabel('Number of Tosses')
plt.ylabel('Probability of Heads')
plt.legend()
plt.show()

print("Final Result after 1000 tosses:", probabilities[-1])
```
