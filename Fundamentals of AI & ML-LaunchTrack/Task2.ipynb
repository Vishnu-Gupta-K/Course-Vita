```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("/content/movie recommendations dataset.csv")

```

```python
# Look at the first few rows
print("Preview of data:")
print(df.head())

# Check general info
print("\nData Info:")
print(df.info())

```

```python
# Check missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# Remove exact duplicate rows
df = df.drop_duplicates()
print("\nAfter removing duplicates:", df.shape)
```

```python
# Clean whitespace in object columns
for col in df.select_dtypes(include="object"):
    df[col] = df[col].astype(str).str.strip().replace({"": np.nan})

print("\nCleaned text columns.")
```

```python
# Example using NumPy for learners
arr = np.array([3.5, np.nan, 4.8, -1, 5.2])

# Replace negative or nan with median
valid = arr[arr >= 0]
median_val = np.median(valid)

arr_fixed = np.where(arr < 0, median_val, arr)
arr_fixed = np.where(np.isnan(arr_fixed), median_val, arr_fixed)

print("\nOriginal array:", arr)
print("Fixed array:", arr_fixed)
```
